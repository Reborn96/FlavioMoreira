## Criando novas variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('Dados/aluguel.csv', sep = ';')
dados.head(10)
```

```python
dados['Valor bruto'] = dados['Valor']+dados['Condominio']+dados['IPTU']#FAzendo contas com colunas
dados.head(10)#adicionar uma coluna é simples, é somente iniciar ela com o 'Nome' e descrever como deve ser preenchida
```

```python
dados['Valor M2'] = dados['Valor']/dados['Area']
dados['Valor M2']
```

```python
dados['Valor M2'] = dados['Valor M2'].round(2)#função .round(x) descreve quantas casas decimais
dados.head(10)
```

```python
dados['Valor bruto m2'] = (dados['Valor bruto'] / dados['Area']).round(2)
dados.head(10)
```

```python
tipos = []#Eu que fiz esse hehe, pra não ter que digitar o tipo de casas
for tipo in dados['Tipo']:
    if tipo not in tipos:
        tipos.append(tipo)
print(tipos)
```

```python
tipos.pop(0)
```

```python
tipos.pop(3)
```

```python
casa = tipos.copy()#copiando pra não ficar vinculado, não era necessário, mas vale pontos de estilo
casa
```

```python
dados.pop('Tipo Agregado')#limpando a base que tava zoada
```

```python
dados.head(10)
```

```python
dados['Tipo agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados#lambda descreve como preencher uma casa dadas algumas condições
```

```python
dados.to_csv('Dados/aluguel_residencial.csv', sep = ';', index = False)
```
