```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize= (20,10))
print('ok')
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head()
```

```python
area = plt.figure()#iniciando a variável do gráfico, por enquanto vazia
```

```python
g1 = area.add_subplot(2,2,1)#Descrevemos a estrutura do gráfico, sendo 2 linhas, 2 colunas, e o gráfico em questão o 1º
g2 = area.add_subplot(2,2,2)#Descrevemos a estrutura do gráfico, sendo 2 linhas, 2 colunas, e o gráfico em questão o 2º
g3 = area.add_subplot(2,2,3)#Descrevemos a estrutura do gráfico, sendo 2 linhas, 2 colunas, e o gráfico em questão o 3º
g4 = area.add_subplot(2,2,4)#Descrevemos a estrutura do gráfico, sendo 2 linhas, 2 colunas, e o gráfico em questão o 4º
```

```python
g1.scatter(dados['Valor'],dados['Area'])#Grafico de dispersão, mostrando os eixos
g1.set_title('Valor x Área')#colcoando titulo

g2.hist(dados['Valor'])#Histograma
g2.set_title('Histograma')

dados_g3 = dados['Valor'].sample(100)#.sample adquire 100 amostras aleatórias
dados_g3.index = range(dados_g3.shape[0])#somente dando que o indice que o tamanho do ds
g3.plot(dados_g3)
g3.set_title('Amostra de valor')

grupo = dados.groupby('Tipo')['Valor']#Agrupa o tipo de imovel a partir de valores
label = grupo.mean().index#descreve quais os tipos avaliados
valores = grupo.mean().values#adquire o valor médio de cada tipo de imóvel da serie grupo
g4.bar(label, valores)
g4.set_title('Valor médio por tipo')

ex1 = grupo.mean().index#descreve quais os tipos avaliados
ex2 = grupo.count().index#descreve quais os tipos avaliados
print(ex1)
print(ex2)
print(valores)
```
