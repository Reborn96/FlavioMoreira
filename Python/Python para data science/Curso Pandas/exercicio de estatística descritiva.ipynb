```python
%matplotlib inline
import pandas as pd
import matplotlib as plt
plt.rc('figure',figsize = (20,10))
print('bibliotecas importadas')
```

```python
dados = pd.read_csv('dados/aluguel_amostra.csv', sep = ';')
dados.head(20)
```

```python
q1 = dados['Valor m2'].quantile(.25)
print(f'quartil 1: {q1}')
q3 = dados['Valor m2'].quantile(.75)
print(f'quartil 3: {q3}')
iiq = (q3 - q1).round(2)
print(f'IIQ: {iiq}')
liminf = (q1 - 1.5 * iiq).round(2)
limsup = (q3 + 1.5 * iiq).round(2)
print(f'Limite inferior {liminf}')
print(f'Limite superior {limsup}')
```
