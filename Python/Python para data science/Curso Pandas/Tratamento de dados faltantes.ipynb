## Tratamento de dados faltantes

```python
import pandas as pd
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dados
```

```python
dados.isnull()#verificar se o DF tem valores nulos
```

```python
dados.notnull()#Mesma coisa, só que inverso
```

```python
dados.info()#muito útil para pegar overwiew do df
```

```python
dados[dados['Area'].isnull()]#método para verificar quais são os registros nulos
```

```python
shape = dados.shape[0]
dados.dropna(subset = ['Area'], inplace = True)#função dropna funciona para se livrar de NaN, necessário adicionar inplace = True
```

```python
shape2 = dados.shape[0]
shape2
```

```python
shape - shape2#registros removidos
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
selecao#selecionando quais apartamentos são nulos e atribuindo variável
```

```python
antes = dados.shape[0]
dados = dados[~selecao]# o ~ faz com que a matriz booleana inverta, como selecionamos como False os que quremos salvar, rpecisamos inverter para livrar do resto
depois = dados.shape[0]
antes - depois
```

```python
dados.fillna(0, inplace = True)#metodo para preencher valores nulos com 0
```

```python
dados.fillna({'Condominio':0, 'IPTU':0})#metodo 2 para preencher valores nulos com 0
```

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';',index = False)#Salvando novo docto, e não salvando o indice
```
