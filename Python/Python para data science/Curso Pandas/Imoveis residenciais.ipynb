## Imoveis residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head(10)
```

```python
tipo = list(dados['Tipo'].drop_duplicates())#importante deixar como lista se não, não funciona
res = ['Casa', 'Casa de Condomínio', 'Casa de Vila', 'Quitinete']
```

```python
sel = dados['Tipo'].isin(res)#Confere todas as entradas 'Tipo' para verificar se está a lista
dados_res = dados[sel]#Criando um DF com os valores True
dados_res.head(10)
```

```python
list(dados_res['Tipo'].drop_duplicates())#removendo duplicatas
```

```python
dados_res.shape[0]#verificando tamanho do DF
```

```python
dados.shape[0]#comparando com original
```

```python
dados_res.index
#index = ['Cód ' + str(i) for i in range(dados_res.shape[0])] #Não é necessário ir tão longe, podemos mudar direto
dados_res.index = range(dados_res.shape[0])#mudando direto
dados_res
```

# Exportando a base de dados

```python
dados_res.to_csv('dados/aluguel_residencial.csv', sep = ';')#to_csv para exportar como csv
print('Dados exportados com sucesso!')#Mantem as coisas um pouco organizadas
```
